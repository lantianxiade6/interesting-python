```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyecharts.charts import Bar,Line
import seaborn as sns
import pylab as mpl

mpl.rcParams['font.sans-serif']=['SimHei']
plt.rc('figure',figsize=(10,10))
%matplotlib inline
```

```python
data=pd.read_csv('NationalNames.csv')
data.head()
data.groupby('Year').Count.sum()
```

```python
#取出Year=2010-2017和Gender=M的行，并按Name汇总，并按Count求和，取出最大的15个
top15_boy=data.loc[(data['Year'].isin(list(range(2010,2018))))&(data['Gender']=='M'),:].groupby('Name').Count.sum().nlargest(15)
top15_boy
```

```python
from wordcloud import WordCloud
value=list(top15_boy.values)
name=list(top15_boy.index)
freq={key:val for key in name for val in value}
pic=WordCloud(height=100,width=100,background_color='#f2eada')
pic.generate_from_frequencies(freq)
mpl.imshow(pic)
```

```python
#取出规定Year=2010-2017和Gender=M下，且名字在top15_boy里面的数据
data_top15_boy = data.loc[(data['Year'].isin(list(range(2010,2018))))&(data['Gender']=='M')&(data['Name'].isin(list(top15_boy.index))),:]
#按Year和count按升序排序
data_top15_boy.sort_values(by=['Year','Count'],ascending=True,inplace=False)
data_top15_boy.head()
```

```python
sns.set(font_scale=1.4)#设置文字大小
g=sns.FacetGrid(data_top15_boy,col="Name",col_wrap=3)#设置每个Name一张图，每行3列
g=g.map(plt.plot,"Year","Count",color="c",marker=".")#横坐标为Year，纵坐标为Count画15个图
```

```python
from pyecharts import options as opts
def name_trend(name,data,Year=1920,dodge=500):
    name_data=data[(data['Name']==name)&(data['Year']>Year)]
    attr=list(range(Year,2018))
    attr=[str(i) for i in attr]
    v1=name_data[name_data['Gender']=='M'].Count.values
    v2=name_data[name_data['Gender']=='F'].Count.values
    v1=[int(i) for i in v1]
    v2=[int(i) for i in v2]
    v1d=[int(i)+dodge for i in v1]
    v2d=[int(i)+dodge for i in v2]
    bar=Bar()
    bar.add_xaxis(attr)
    bar.add_yaxis('男',v1,label_opts=opts.LabelOpts(is_show=False))
    bar.add_yaxis('女',v2,label_opts=opts.LabelOpts(is_show=False))
    
    line=Line()
    line.add_xaxis(attr)
    line.add_yaxis('男',v1d,label_opts=opts.LabelOpts(is_show=False))
    line.add_yaxis('女',v2d,label_opts=opts.LabelOpts(is_show=False))
    
    bar.overlap(line)
    return bar.render_notebook()
name_trend('Jordan',data)
```
